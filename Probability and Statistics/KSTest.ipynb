```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
'''
- I'm generating x with 1000 samples using stats.norm.random variables. generating 1000 size of arrays and all 
these values are gaussian distributed using normal distribution.
- Basically X ~ N(0,1)
- setup grid wise
- kernel density estimation plots the PDF and I'm justing numpy array x
'''
#generating a gaussian random variable x

x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw_method=0.5) # bw_method; method for determining smoothing bandwidth to use
plt.show()
```

```python
'''
the above plot looks like gaussian dist. mean of 0 and variance of 1
'''

'''
KSTest; to determine whether our values in x (1000 values) follows gaussian distribution or not for this: use kstest() method
'''
stats.kstest(x, 'norm') 
```

```python
''' 
give with x and compare with normal distribution. As soon as we run, we got two value Statistic (d) and p value
here our p-value is quite high and hence random variable X is distributed normally
'''
```

```python
#Y - Continuous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000) #generating 10000 samples of from this random variable and storing on y
sns.kdeplot(np.array(y), bw_method=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

```python
'''
Our p-value becomes 0, as a result, our y is not normally distributed
'''
```
