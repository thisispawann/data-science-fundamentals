Q-Q Plot : How to test if a random variable is normally distributed or not?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# N(0,1)
# this is how we generate a sample of random number
# this is how we got observations from normal distribution

std_normal = np.random.normal(loc=0, scale=1, size=1000) # loc 0 is mean (mu), scale=1 is s.d
''' above lines code says, generate random numbers or sample randomly from Gaussian/normal distribution with mean 0,
s.d 1 and 1000's of such samples and put all 1000 variables in std_normal variable. This is how we generate random
observations from Gaussian distribution.'''

# 0 to 1000th percentiles of std_normal
for row in range(0, 101):
    print(row, np.percentile(std_normal, row))
```

```python
# generate 100 samples from N(20,5)

measurements = np.random.normal(loc=20,scale=5, size=100)

'''statistics module in numpy has a probplot function. here we can say we want a Q-Q plot between my variable measurement
and i want to compare with normal (compare with N(0,1) and potting using the library plt)'''
stats.probplot(measurements, dist='norm', plot=plt)
plt.show()
```

Here our X-axis is N(0,1) and y-axis is N(20,5). They are coming from same gaussian family. they have different mean and s.d but have same type of distribution both X and Y are gaussian. Now, instead of taking 100 samples let's take 1000 samples

```python
measurements = np.random.normal(loc=20,scale=5, size=1000)
stats.probplot(measurements, dist='norm', plot=plt)
plt.show()
```

when i made it 100 to 1000 samples, more of the points falls under this red lines except the extremes ones.

```python
measurements = np.random.normal(loc=20,scale=5, size=50000)

stats.probplot(measurements, dist='norm', plot=plt)
plt.show()
```

the moment we make it 50000 as my number of samples increases more and more points are closer to red lines. whenever we have a small size of samples/observation (such as 50), Q-Q plot will not work. this is the major limitations of Q-Q plot.

```python
'''
lets change the normal distribution to uniform distribution and analyze the plot
'''
measurements = np.random.uniform(low=-1, high=1, size=1000)
stats.probplot(measurements, dist='norm', plot=plt)
plt.show()
```

In my x-axis, i have Gaussian distribution and in y-axis, i have a uniform distribution. the point in the plot is going farther and farther from the red line. it means is extremity of this distribution is diverging more and more

```python
measurements = np.random.uniform(low=-1, high=1, size=10000) #lets take 10000 samples
stats.probplot(measurements, dist='norm', plot=plt)
plt.show()
```

there is significantly deviation now, more and more points are deviating from the line. if the points are lie on the line it concludes x and y axis are on the same family.

Application oi Q-Q plot:

as we saw Q-Q plot as method to give an answer these questions:

1. is X ~ N(mu, sd) [is x is normally distributed?]
2. if we have a random variable x and y, does x and y have the same distribution?

Therefore, Q-Q plot is a graphical technique just to give an answer of these two questions.
