Matplotlib:

- It is a 2D plotting library for Python Programming that is used for data visualization to generalize various figures and high-quantity image plots; with the help of the Matplot library, we can easily generalize the various diagrams like histograms, scatter plots, plt error charts, bar charts with just only a few lines of code.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,6,4]) # y axis
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

Plot with grid

```python
plt.plot([1,2,3,4],[1,4,9,16]) # x-axis and y-axis
plt.ylabel('Squares')
plt.xlabel('numbers')
plt.grid()

plt.show()
```

```python
t = np.arange(0., 5., 0.2) #start, end, steps

#blue dashes, red squares, green triangles
plt.plot(t, t**2, 'b--', label = '^2')
plt.plot(t, t**2.2, 'rs', label = '^2.2')
plt.plot(t, t**2.5, 'g^', label = '^2.5')

plt.grid()

plt.legend()
plt.show()
```
