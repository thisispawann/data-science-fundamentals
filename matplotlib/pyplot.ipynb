Matplotlib:

- It is a 2D plotting library for Python Programming that is used for data visualization to generalize various figures and high-quantity image plots; with the help of the Matplot library, we can easily generalize the various diagrams like histograms, scatter plots, plt error charts, bar charts with just only a few lines of code.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,6,4]) # y axis
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

Plot with grid

```python
plt.plot([1,2,3,4],[1,4,9,16]) # x-axis and y-axis
plt.ylabel('Squares')
plt.xlabel('numbers')
plt.grid()

plt.show()
```

```python
t = np.arange(0., 5., 0.2) #start, end, steps

#blue dashes, red squares, green triangles
plt.plot(t, t**2, 'b--', label = '^2')
plt.plot(t, t**2.2, 'rs', label = '^2.2')
plt.plot(t, t**2.5, 'g^', label = '^2.5')

plt.grid()

plt.legend()
plt.show()
```

using set property; setp()

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]

x2 = [1,2,3,4]
y2 = [2,4,6,8]

lines = plt.plot(x1, y1, x2, y2)

#use keyword args
plt.setp(lines[0], color='r', linewidth = 2.0) # lines 0 --> x1, y1 line
plt.setp(lines[1], color='g', linewidth = 2.0) # lines 0 --> x2, y2 line
plt.grid()
```

Break a single figure plot into parts

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()

plt.subplot(211) # 2 -> rows, 1-> column, 1->figure no.
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212) # 2 -> rows, 1-> column, 1->figure no.
plt.plot(t2, f(t2), 'r--')
plt.show()
```
