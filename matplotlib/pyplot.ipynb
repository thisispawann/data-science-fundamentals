Matplotlib:

- It is a plotting library. Each pyplot function makes some changes to figure: such as, creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([2,3,6,4])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```
